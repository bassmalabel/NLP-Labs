```python
import gensim.downloader as api
word2vec = api.load("word2vec-google-news-300")  # downloads automatically
word2vec.most_similar('king', topn=5)
```

```python
from gensim.models import KeyedVectors

word2vec = KeyedVectors.load_word2vec_format(
    'GoogleNews-vectors-negative300.bin', 
    binary=True
)

# Get 5 most similar words
word2vec.most_similar('king', topn=5)
```

```python
glove = KeyedVectors.load_word2vec_format(
    'glove.6B.300d.word2vec.txt', 
    binary=False
)

glove.most_similar('king', topn=5)
```

```python
fasttext = KeyedVectors.load_word2vec_format(
    'cc.en.300.vec', 
    binary=False
)

fasttext.most_similar('king', topn=5)
```
